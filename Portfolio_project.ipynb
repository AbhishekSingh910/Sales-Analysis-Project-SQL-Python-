```python
# by default it considers few values as NA, but I want to it 
# consider Not-Available, unkowon as NA only

df1 = pd.read_csv('orders_data.csv',na_values=['Not Available','unknown'])
df1['Ship Mode'].unique()
```

#**Derive new columns discount , sale price and profit**

```python
df1.head(2)
```

```python
#Selling Pirce
df1['Selling_Price'] = df1['List Price'] - df1['List Price']*(df1['Discount Percent']/100)
df1['Selling_Price']
```

```python
df1.head(1)
```

```python
#PROFIT 
df1['Profit'] = df1['Selling_Price'] - df1['cost price']
df1['Profit']

```

```python
df1.head(4)
```

```python
df1.rename(columns={'Profit':'Unit_Profit', 'Selling_Price':'Unit_Selling_Price'},inplace=True)
df1.head()
```

```python
df1.dtypes
```

## *changing datatype of Order Date column from object to datetime*

```python
df1['Order Date'] = pd.to_datetime(df1['Order Date'], format="%d-%m-%y")
df1
```

```python
df1.dtypes
```

## *Dropping few columns*

```python
df1.drop(columns = {'cost price','List Price', 'Discount Percent'},inplace = True)
```

```python
df1.head(3)
```

```python
# Calculating total profit in each Order

df1['Total Profit'] = df1['Quantity'] * df1['Unit_Profit']
df1.head()
```

### Q14 Find month over month growth comparison for 2022 and 2023 sales eg : jan 2022 vs jan 2023

```python
df1.head(2)
```

```python
a =df1.groupby(['Year','Month'])['Total_sales'].sum().reset_index()
a
```

```python
pivot_table = a.pivot_table(index='Month', columns='Year', values='Total_sales', aggfunc='sum').reset_index()
pivot_table
```

```python
pivot_table['YoY_growth'] = ((pivot_table[2023] - pivot_table[2022]) / pivot_table[2022] )*100
pivot_table
```
